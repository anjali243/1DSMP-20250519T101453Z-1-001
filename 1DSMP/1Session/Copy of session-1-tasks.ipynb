### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
x1_str,y1_str=input("Enter coordinates of point A (x1,y1)").split()

x2_str,y2_str=input("Enter coordinates of point B (x2,y2)").split()
x1=float(x1_str)
y1=float(y1_str)
x2=float(x2_str)
y2=float(y2_str)


Euclidian_distance= ((x2-x1)**2 + (y2-y1)**2)**0.5
print("Euclidian_distance:", Euclidian_distance)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
heads=int(input("Enter the number of total heads: "))
legs=int(input("Enter the number of total legs: "))
chicken=(4*heads-legs)//2
dogs=heads-chicken
print("Chicken",chicken,"Dogs",dogs)

````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
number=int(input("enter a number"))
sum=0
for i in range(1,number+1):
  sum=sum+i*i

print("Sum of squares of first", number, "natural numbers is:", sum)
````

````python
# Input: number of natural numbers
n = int(input("Enter the value of n: "))

# Formula-based sum of squares: n(n+1)(2n+1)/6
sum_of_squares = n * (n + 1) * (2 * n + 1) // 6

print("Sum of squares of first", n, "natural numbers is:", sum_of_squares)
````
